```python
import pandas as pd
from scipy.stats import norm
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
```

# Chargement

```python
df = pd.read_csv('Données clients - Feuille 1_nettoyées.csv')
```

# Distribution du budget

```python
plt.figure(figsize=(10, 6))
plt.hist(df['budget_voyage_annuel'], bins=10, density=True, color='skyblue', edgecolor='black', alpha=0.6)

mu, std = norm.fit(df['budget_voyage_annuel'])
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, df.shape[0])
p = norm.pdf(x, mu, std)

plt.plot(x, p, 'k', linewidth=2, label="Ajustement gaussien")

plt.legend()

plt.title('Distribution du budget annuel des clients')
plt.xlabel('Budget')
plt.ylabel('Densité de clients')
plt.grid(axis='y', alpha=0.75)
```

```python
mean_budget = df['budget_voyage_annuel'].mean()
std_budget = df['budget_voyage_annuel'].std()

print(f"Moyenne du budget annuel : {round(mean_budget, 0)}")
print(f"Écart type du budget annuel : {round(std_budget, 0)}")
```
