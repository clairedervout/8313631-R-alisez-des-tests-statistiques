```python
import pandas as pd
import seaborn as sns
```

# Test de Spearman

```python
from scipy.stats import spearmanr
```

```python
df = pd.read_csv('/content/Données clients - Feuille 1_nettoyées.csv')
```

```python
df.head()
```

```python
sns.regplot(data=df, x="durée de voyage (en jours)", y="prix total", robust=True, line_kws=dict(color="r"))
```

```python
spearman_corr, spearman_p_value = spearmanr(df['durée de voyage (en jours)'], df['prix total'])

print(f"Coefficient de corrélation de Spearman: {spearman_corr}")
print(f"Valeur p: {spearman_p_value}")
```
