```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
```

# Chargement

```python
df = pd.read_csv('/content/Données clients - Feuille 1_nettoyées.csv')
```

# Distribution des âges

```python
plt.figure(figsize=(10, 6))
plt.hist(df['prix total'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution du Prix Total des Voyages')
plt.xlabel('Prix Total')
plt.ylabel('Nombre de Voyages')
plt.grid(axis='y', alpha=0.75)
```

```python
print("La médiane des âges est de :", df["âge"].median(), "ans.")
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(df['âge'])
plt.title('Distribution des âges des clients')
plt.xlabel('âge')
plt.ylabel('Nombre de clients')
plt.grid(axis='y', alpha=0.75)
```
