#A vous de jouer

H0 = il n’y a pas de corrélation linéaire entre les deux variables

H1 = il y a une corrélation linéaire entre les deux variables

```python
from scipy.stats import pearsonr
```

```python
df = pd.read_csv('/content/Données entreprise Vertigo nettoyées - Feuille 1.csv')
```

```python
df.head()
```

```python
sns.regplot(data=df, x="Nombre_de_Clients", y="Revenu_des_Ventes", robust=True, line_kws=dict(color="r"))
```

```python
correlation, p_value = pearsonr(df["Nombre_de_Clients"], df["Revenu_des_Ventes"])

print(f"Coefficient de corrélation de Pearson: {correlation}")
print(f"Valeur p: {p_value}")
```

On peut donc rejetté H0 et conclure qu'il y a une corrélation linéaire entre les deux variables.
