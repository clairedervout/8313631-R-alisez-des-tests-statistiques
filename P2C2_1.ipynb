# Value P: explications

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Paramètres de la distribution normale
mu = 0      # Moyenne
sigma = 1   # Ecart type

# Génération de 1000 valeurs suivant une distribution normale
samples = np.random.normal(mu, sigma, 1000)

# Tracé de l'histogramme des échantillons
plt.hist(samples, bins=30, edgecolor='black', alpha=0.7, density=True)

# Ajout de la courbe de la distribution normale pour comparaison
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = (1/(sigma * np.sqrt(2 * np.pi))) * np.exp( - (x - mu)**2 / (2 * sigma**2))
plt.plot(x, y, 'r-', label='Loi normale')

plt.title('Histogramme des échantillons suivant une loi normale')
plt.xlabel('Valeurs')
plt.ylabel('Densité')
plt.legend()
plt.grid(True)
plt.show()
```

# Les données vertigo

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm

# Charger les données depuis le fichier CSV
file_path = '/content/Données clients - Feuille 1_nettoyées.csv'
data = pd.read_csv(file_path)
```

# Distribution du budget annuel par client

```python
# Extraire la colonne 'budget_voyage_annuel'
budgets = data['budget_voyage_annuel'].dropna()

plt.figure(figsize=(10, 6))
plt.hist(budgets, bins=10, density=True, color='skyblue', edgecolor='black', alpha=0.6)

mu, std = norm.fit(budgets)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, len(budgets))
p = norm.pdf(x, mu, std)

plt.plot(x, p, 'k', linewidth=2, label="Ajustement gaussien")

plt.legend()

plt.title('Distribution du budget annuel des clients')
plt.xlabel('Budget')
plt.ylabel('Densité de clients')
plt.grid(axis='y', alpha=0.75)
```

# Test de normalité

```python
from scipy.stats import shapiro

# Effectuer le test de Shapiro-Wilk
stat, p_value = shapiro(budgets)

# Afficher les résultats
print("Taille de mon échantillon:", len(budgets))
print(f"Statistique du test de Shapiro-Wilk : {stat}")
print(f"Valeur p : {p_value}")

# Interprétation des résultats
alpha = 0.05
if p_value > alpha:
    print("Les données suivent une distribution normale (on ne rejette pas H0)")
else:
    print("Les données ne suivent pas une distribution normale (on rejette H0)")
```
