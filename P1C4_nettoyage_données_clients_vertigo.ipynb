# Les outliers

```python
# Colonnes quantitative
quanti_col = ["âge", "durée de voyage (en jours)",	"nombre de personnes",	"prix total", "évaluation sur 5", "budget_voyage_annuel", "durée moyenne de voyage annuelle (en jours)"]
# quanti_col = ["Nombre_de_Clients", "Revenu_des_Ventes", "Depenses_Publicitaires", "Nombre_de_Reservations"]
```

```python
# Méthode de détection par la distance interquartile.
def detect_outliers(df, feature):
  Q1 = df[feature].quantile(0.25)
  Q3 = df[feature].quantile(0.75)
  IQR = Q3 - Q1

  return (df[feature] < (Q1 - 2 * IQR)) | (df[feature] > (Q3 + 2 * IQR))
```

```python
df[detect_outliers(df, "âge")]
# df[detect_outliers(df, "Nombre_de_Clients")]
```

```python
# Supprimer les outliers
for col in quanti_col:
  if col in df.columns:
    df = df[~detect_outliers(df, col)]
```

# Valeurs manquantes

```python
# On compte
df.isnull().sum()
```

```python
# On visualise
df[df.isnull().any(axis=1)]
```

```python
# On remplace par la médiane
values = {
    c: df[c].median() if df[c].dtype != 'object' else df[c].mode()[0] for c in df.columns
}
df.fillna(value=values, inplace=True)
```
